```python
import json
import pandas as pd

with open("all_Lin_Shu_token_pos_ner_by_subfolder.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data)
print(df.columns)
```

```python
def lexical_features(tokens):
    word_count = len(tokens)
    type_count = len(set(tokens))
    ttr = type_count / word_count if word_count > 0 else 0
    avg_len = sum(len(t) for t in tokens) / word_count if word_count > 0 else 0
    return pd.Series({'word_count': word_count, 'type_token_ratio': ttr, 'avg_token_len': avg_len})

df_lex = df['tokens'].apply(lexical_features)
df = pd.concat([df, df_lex], axis=1)
```

```python
from collections import Counter

def pos_distribution(pos_list):
    count = Counter(pos_list)
    total = sum(count.values())
    return pd.Series({f'POS_{tag}': count[tag]/total for tag in count})

df_pos = df['pos'].apply(pos_distribution).fillna(0)
df = pd.concat([df, df_pos], axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

df['text_str'] = df['tokens'].apply(lambda x: ' '.join(x))

vectorizer = CountVectorizer(max_features=100)  # change to top-N MFW
X_mfw = vectorizer.fit_transform(df['text_str'])
df_mfw = pd.DataFrame(X_mfw.toarray(), columns=[f'MFW_{w}' for w in vectorizer.get_feature_names_out()])
df = pd.concat([df, df_mfw], axis=1)
```

```python
from scipy.stats import ttest_ind

def compare_groups(feature):
    group_labels = df['group'].unique()
    if len(group_labels) != 2:
        return None  # Only works for 2 groups for now
    a = df[df['group'] == group_labels[0]][feature]
    b = df[df['group'] == group_labels[1]][feature]
    stat, p = ttest_ind(a, b, equal_var=False)
    return pd.Series({'groupA': group_labels[0], 'groupB': group_labels[1], 't_stat': stat, 'p_value': p})

features_to_compare = ['word_count', 'type_token_ratio', 'avg_token_len'] + \
                      list(df_pos.columns) + list(df_mfw.columns)

results = pd.DataFrame([compare_groups(f) for f in features_to_compare], index=features_to_compare)
significant = results[results['p_value'] < 0.05]
print(significant)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Combine all stylometric features
feature_cols = ['word_count', 'type_token_ratio', 'avg_token_len'] + \
               list(df_pos.columns) + list(df_mfw.columns)

X = df[feature_cols].fillna(0)
X_scaled = StandardScaler().fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

df['PC1'], df['PC2'] = X_pca[:, 0], X_pca[:, 1]

# Plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='PC1', y='PC2', hue='group', s=100)
plt.title('PCA of Stylometric Features')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)')
plt.grid(True)
plt.show()
```

```python
# Show group means for key features
summary_features = ['word_count', 'type_token_ratio', 'avg_token_len']
summary = df.groupby("group")[summary_features].mean().round(2)
print("\n📊 Group-level Lexical Summary:")
print(summary)

# Optional: save to CSV
summary.to_csv("group_lexical_summary.csv")
```

```python
import json
import pandas as pd

# === Load your JSON file ===
with open("all_Lin_Shu_token_pos_ner_by_subfolder.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# === Parameters ===
chunk_size = 2000

# === Chunking process ===
chunked_rows = []

for entry in data:
    tokens = entry['tokens']
    pos_tags = entry['pos']
    group = entry['group']
    filename = entry['filename']

    # Sanity check
    assert len(tokens) == len(pos_tags)

    # Split into 2000-token chunks
    for i in range(0, len(tokens), chunk_size):
        token_chunk = tokens[i:i+chunk_size]
        pos_chunk = pos_tags[i:i+chunk_size]

        if len(token_chunk) < 100:  # skip tiny trailing chunk if needed
            continue

        chunked_rows.append({
            'filename': f"{filename}_chunk{i//chunk_size+1}",
            'group': group,
            'tokens': token_chunk,
            'pos': pos_chunk,
            'chunk_id': i // chunk_size + 1,
            'chunk_len': len(token_chunk)
        })

# === Create chunked DataFrame ===
df_chunks = pd.DataFrame(chunked_rows)
print(f"✅ Created {len(df_chunks)} chunks from {len(data)} texts.")

# === Optional: Save to JSON or CSV ===
df_chunks.to_json("chunked_2000_token_blocks.json", force_ascii=False, orient="records", indent=2)
df_chunks.to_csv("chunked_2000_token_blocks.csv", index=False, encoding="utf-8-sig")
```

```python
function_words = ['之', '以', '而', '者', '乃', '為', '爾', '矣', '則', '卽']
```

```python
from collections import Counter

def function_word_freq(tokens):
    total = len(tokens)
    count = Counter(tok for tok in tokens if tok in function_words)
    freqs = {f'FW_{w}': count[w]/total if total > 0 else 0 for w in function_words}
    return pd.Series(freqs)

df_func = df['tokens'].apply(function_word_freq)
df = pd.concat([df, df_func], axis=1)
```

```python
group_means = df.groupby("group")[[f'FW_{w}' for w in function_words]].mean().T
group_means.columns.name = None
group_means = group_means.round(5)
print("📊 Function Word Frequency by Group (per 2000-token chunk):")
print(group_means)
```

```python
# Step 1: Compute relative frequency of pronouns in each chunk
def pronoun_freq(pos_list):
    count = sum(1 for tag in pos_list if tag == 'PN')
    return count / len(pos_list) if len(pos_list) > 0 else 0

df['pronoun_rel_freq'] = df['pos'].apply(pronoun_freq)

# Step 2: Compute group-level mean
pronoun_group_means = df.groupby('group')['pronoun_rel_freq'].mean().round(5)
print("📊 Average Relative Frequency of Pronouns (代词) by Group:")
print(pronoun_group_means)
```

```python
import matplotlib.pyplot as plt

pronoun_group_means.plot(kind='bar', color='skyblue', figsize=(6, 4))
plt.title("Relative Frequency of Pronouns (代词) per 2000-token Chunk")
plt.ylabel("Relative Frequency")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from collections import Counter

# List of pronouns of interest
target_pronouns = ["吾", "余", "此", "其", "人", "我", "爾", "汝", "女", "彼", "王", "君", "衆"]

# Step 1: Compute relative frequency of each pronoun per chunk
def count_target_pronouns(tokens):
    total = len(tokens)
    counter = Counter(tokens)
    return {f'PRON_{w}': counter[w]/total if total > 0 else 0 for w in target_pronouns}

df_pron = df['tokens'].apply(count_target_pronouns).apply(pd.Series)
df = pd.concat([df, df_pron], axis=1)

# Step 2: Group-level means
group_pron_means = df.groupby("group")[[f'PRON_{w}' for w in target_pronouns]].mean().T.round(5)
group_pron_means.columns.name = None
print("📊 Relative Frequency of Selected Pronouns per Group:")
print(group_pron_means)
```

```python
import matplotlib.pyplot as plt

group_pron_means.plot(kind='bar', figsize=(12, 6))
plt.title("Relative Frequency of Selected Pronouns per Group (2000-token chunks)")
plt.ylabel("Relative Frequency")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(title="Group")
plt.tight_layout()
plt.show()
```
